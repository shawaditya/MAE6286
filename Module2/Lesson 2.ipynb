```python
import numpy
from matplotlib import pyplot

%matplotlib inline
```

```python
def linear_convection(nx, L=2,c=1.0, dt=0.025, nt=20):
    dx   = L / (nx-1)
    x    = numpy.linspace(0.0,L,num=nx)
    u0   = numpy.ones(nx)
    mask = numpy.where((x>=0.5)&(x<=1.0))
    u0[mask]  = 2.0
    u = u0.copy()
    for n in range (nt):
        un = u.copy()
        u[1:] = un[1:] - c*dt/dx*(un[1:] - un[:-1])
    pyplot.figure()
    pyplot.grid()
    pyplot.plot(x, u0, label='Initial')
    pyplot.plot(x, u, label='Current')
    pyplot.legend()
    pyplot.xlim(0.0,L)
    pyplot.ylim(0.0,2.5)
    
```

```python
linear_convection(41)
```

```python
linear_convection(61)
```

```python
linear_convection(81)
```

```python
linear_convection(81,nt=10)
```

## CFL

here,
     sigma = c * dt/dx
     
Thus we need to fix sigma instead of dt and compute.

```python
def linear_convection_cfl(nx, L=2,c=1.0, sigma = 0.5, nt=20):
    dx   = L / (nx-1)
    x    = numpy.linspace(0.0,L,num=nx)
    dt   = dx * sigma / c 
    u0   = numpy.ones(nx)
    mask = numpy.where((x>=0.5)&(x<=1.0))
    u0[mask]  = 2.0
    u = u0.copy()
    for n in range (nt):
        un = u.copy()
        u[1:] = un[1:] - c*dt/dx*(un[1:] - un[:-1])
    pyplot.figure()
    pyplot.grid()
    pyplot.plot(x, u0, label='Initial')
    pyplot.plot(x, u, label='Current')
    pyplot.legend()
    pyplot.xlim(0.0,L)
    pyplot.ylim(0.0,2.5)
    
```

```python
linear_convection_cfl(85)
```

sigma(max) = 1.0
For this condition sigma value is greater than 1. So it blows up.

```python
linear_convection_cfl(85,sigma=1.1)   
```
