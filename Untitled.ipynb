# Eulers Method

```python
import numpy
from matplotlib import pyplot 
%matplotlib inline
```

```python
# Creating the timeline
"""
T is the time interval
dt is the small time difference
N is the number of time steps
"""
T = 200 
dt = 0.01
N = 1 + int(T/dt)
"""
Now we need to create a linspace for time
"""
t = numpy.linspace(0.0,T, num=N)
```

```python
# defining the initial properties
"""
z_int : initial altitude
b_int : initial trim velocity at the moment of perturbation
g : acc. due to gravity
"""
z_int = 20
b_int = 50
ztrim = 17
g = 9.81

# we need to create an array for u vector with intial conditions of z_int and b_int

u = numpy.array([z_int,b_int])

z = numpy.zeros(N)
z[0]=u[0]

# writing the numerical solution

for n in range(1,N):
    func_u = numpy.array([u[1], g*(1-u[0]/ztrim)])
    u = u + dt*func_u
    z[n]=u[0]
    
# writing the exact solution

z_exact = (b_int * (ztrim / g)**0.5 * numpy.sin((g / ztrim)**0.5 * t) +
           (z_int - ztrim) * numpy.cos((g / ztrim)**0.5 * t) + ztrim)

pyplot.plot(t,z_exact,label='exact solution',linestyle='--', color='r')
pyplot.xlabel('t')
pyplot.ylabel('z_exact')
pyplot.plot(t,z,label='numerical solution',linestyle='-',color='y')
pyplot.ylabel('z')
```
